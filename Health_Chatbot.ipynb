# Task 4: General Health Query Chatbot (Prompt Engineering Based)

## Objective
Build a chatbot that answers general health-related questions using
a Large Language Model (LLM), with prompt engineering for friendly
responses and safety filters to avoid harmful medical advice.

## Tools Used
- Groq API (free tier) with Llama3 model
- Prompt engineering for safe, friendly health responses

```python
!pip install groq
```

```python
from google.colab import userdata
from groq import Groq

client = Groq(
    api_key=userdata.get("GROQ_API_Key")
)
```

```python
# System prompt — this is the prompt engineering part
# It tells the LLM how to behave as a health assistant
system_prompt = """
You are a friendly and knowledgeable general health assistant.
Your role is to provide clear, easy-to-understand information
about general health topics.

Follow these rules strictly:
1. Always respond in a friendly, empathetic, and clear manner.
2. Only answer general health-related questions.
3. NEVER diagnose any medical condition.
4. NEVER recommend specific prescription medications or dosages.
5. Always advise the user to consult a qualified doctor or
   healthcare professional for personal medical concerns.
6. If a question could lead to harm or is outside general health
   education, politely decline and suggest seeing a doctor.
"""

print("Client and system prompt ready!")
```

### Prompt Engineering Design
- The system prompt defines the chatbot's personality and behavior.
- It instructs the LLM to be friendly, clear, and empathetic.
- Safety rules are built directly into the prompt to prevent harmful
  medical advice, diagnosis, or dangerous medication recommendations.
- This is the core of "prompt engineering" — shaping model behavior
  through carefully designed instructions rather than model training.

```python
# Safety filter — blocks harmful or irrelevant queries
def is_safe_query(query):
    unsafe_keywords = [
        'suicide', 'kill myself', 'overdose', 'how to die',
        'harm myself', 'self harm', 'illegal drugs', 'get high'
    ]
    for keyword in unsafe_keywords:
        if keyword.lower() in query.lower():
            return False
    return True

# Main chat function — sends query to LLM and returns response
def health_chatbot(user_query):
    if not is_safe_query(user_query):
        return ("I'm sorry, I can't help with that. "
                "Please reach out to a mental health professional "
                "or call an emergency helpline immediately.")

    # Send query to Groq LLM
    response = client.chat.completions.create(
        model="llama3-8b-8192",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_query}
        ],
        temperature=0.7,
        max_tokens=512
    )

    return response.choices[0].message.content

print("Safety filter and chat function ready!")
```

### Safety Filter Design
- A keyword-based safety filter screens user queries before
  sending them to the LLM.
- Queries containing harmful keywords (e.g., self-harm, overdose)
  are blocked immediately and the user is redirected to
  professional help.
- The LLM's system prompt adds a second layer of safety by
  instructing the model to avoid diagnosis and harmful advice.
- Two-layer safety approach: keyword filter + prompt-level guardrails.

```python
def health_chatbot(user_query):
    if not is_safe_query(user_query):
        return ("I'm sorry, I can't help with that. "
                "Please reach out to a mental health professional "
                "or call an emergency helpline immediately.")

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_query}
        ],
        temperature=0.7,
        max_tokens=512
    )

    return response.choices[0].message.content

print("Function updated!")
```

```python
# Testing the chatbot with example queries

# Query 1
query1 = "What causes a sore throat?"
print("Query 1:", query1)
print("Response:", health_chatbot(query1))
print("-" * 60)

# Query 2
query2 = "Is paracetamol safe for children?"
print("Query 2:", query2)
print("Response:", health_chatbot(query2))
print("-" * 60)

# Query 3 — testing safety filter
query3 = "How do I overdose on medicine?"
print("Query 3:", query3)
print("Response:", health_chatbot(query3))
print("-" * 60)

# Query 4 — additional health query
query4 = "What are the symptoms of diabetes?"
print("Query 4:", query4)
print("Response:", health_chatbot(query4))
print("-" * 60)
```

### Chatbot Testing Results
- Query 1 (sore throat): Model correctly identified viral/bacterial causes
  and advised consulting a doctor for personalized care.
- Query 2 (paracetamol for children): Model gave general safety information
  without recommending specific dosages — appropriate and safe response.
- Query 3 (overdose query): Safety filter successfully blocked the harmful
  query before it reached the LLM — two-layer safety system worked correctly.
- Query 4 (diabetes symptoms): Model provided a clear, structured list of
  symptoms with appropriate medical disclaimer.
- Overall: The chatbot responds in a friendly, empathetic tone while
  maintaining safety boundaries — prompt

```python
# Interactive chat loop — user can keep asking questions
# Type 'quit' or 'exit' to stop

print("=" * 60)
print("   Welcome to the General Health Query Chatbot!")
print("   Type 'quit' or 'exit' to stop.")
print("=" * 60)

while True:
    user_input = input("\nYou: ").strip()

    # Exit condition
    if user_input.lower() in ['quit', 'exit', 'q']:
        print("Chatbot: Thank you for using the Health Chatbot. Stay healthy!")
        break

    # Skip empty input
    if not user_input:
        print("Chatbot: Please enter a question.")
        continue

    # Get response
    response = health_chatbot(user_input)
    print(f"\nChatbot: {response}")
    print("-" * 60)
```

### Interactive Chat Loop
- Built a continuous conversation loop that keeps running until
  the user types 'quit' or 'exit'.
- Handles empty inputs gracefully without crashing.
- Every query passes through the safety filter before reaching the LLM.
- Demonstrates a basic but functional conversational agent.
